```python
import pandas as pd
```

```python
df['New Cases'] = df['New Cases'].fillna(0)
df['New Deaths'] = df['New Deaths'].fillna(0)
df['New Recovered'] = df['New Recovered'].fillna(0)
df['New Active Cases'] = df['New Active Cases'].fillna(0)
df['Total Cases'] = df['Total Cases'].fillna(0)
df['Total Deaths'] = df['Total Deaths'].fillna(0)
df['Total Recovered'] = df['Total Recovered'].fillna(0)
df['Total Active Cases'] = df['Total Active Cases'].fillna(0)
df['City or Regency'] = df['City or Regency'].fillna(0)
df['Total Regencies'] = df['Total Regencies'].fillna(0)
df['Total Cities'] = df['Total Cities'].fillna(0)
df['Total Districts'] = df['Total Districts'].fillna(0)
df['Total Urban Villages'] = df['Total Urban Villages'].fillna(0)
df['Total Rural Villages'] = df['Total Rural Villages'].fillna(0)
df['Area (km2)'] = df['Area (km2)'].fillna(0)
df['Population'] = df['Population'].fillna(0)
df['Population Density'] = df['Population Density'].fillna(0)
df['Longitude'] = df['Longitude'].fillna(0)
df['Latitude'] = df['Latitude'].fillna(0)
df['New Cases per Million'] = df['New Cases per Million'].fillna(0)
df['Total Cases per Million'] = df['Total Cases per Million'].fillna(0)
df['New Deaths per Million'] = df['New Deaths per Million'].fillna(0)
df['Total Deaths per Million'] = df['Total Deaths per Million'].fillna(0)
df['Total Deaths per 100rb'] = df['Total Deaths per 100rb'].fillna(0)
df['Growth Factor of New Cases'] = df['Growth Factor of New Cases'].fillna(0)
df['Growth Factor of New Deaths'] = df['Growth Factor of New Deaths'].fillna(0)
```

```python
df.corr()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

korelasi = df.corr()
plt.figure(figsize=(15, 10))
sns.heatmap(korelasi[(korelasi >= 0.5) | (korelasi <= -0.4)], cmap='viridis', vmax=1.0, vmin=-1.0, linewidths=0.1, 
            annot=True, annot_kws={"size": 8}, square=True).set_title('Korelasi Antar Variabel')
plt.show()
```

```python
case_total = df.groupby("Location")["Total Cases per Million"].sum().sort_values()
case_total.plot(kind="barh", fontsize=10, figsize=(8,40), title='Total Kasus yang terjadi di Indonesia')
plt.show()

death_total = df.groupby("Location")["Total Deaths per Million"].sum().sort_values()
death_total.plot(kind="barh", fontsize=10, figsize=(8,40), title='Jumlah Kematian')
plt.show()
```

```python
sns.boxplot(df['Total Deaths per 100rb'])
```

```python
# Scatter plot
fig, ax = plt.subplots(figsize = (20,20))
ax.scatter(df['Total Cases per Million'], df['Total Deaths per Million'])

# x-axis label
ax.set_xlabel('Jumlah Kasus yang terjadi')
 
# y-axis label
ax.set_ylabel('Jumlah Kematian')
plt.show()
```
